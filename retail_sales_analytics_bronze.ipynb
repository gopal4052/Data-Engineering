# Bronze ingest

```python
dbutils.widgets.text(
    "FILE_CHECK_ARRIVAL",
    "",
    "File Arrival Status"
)

file_status = dbutils.widgets.get("FILE_CHECK_ARRIVAL")

if file_status != "FILE_PRESENT":
    raise Exception("File not present")



df_orders = spark.read.format("csv").option("header", "true").option("inferschema","true").load("/Volumes/my_local_data/raw_data/retail_sales_analytics_data/orders.csv")

df_customers = spark.read.format("csv").option("header", "true").option("inferschema","true").load("/Volumes/my_local_data/raw_data/retail_sales_analytics_data/customers.csv")

df_orders.write.format("delta").mode("overwrite").save("/Volumes/my_local_data/bronze/retail_sales/orders")

df_customers.write.format("delta").mode("overwrite").save("/Volumes/my_local_data/bronze/retail_sales/customers")
```
