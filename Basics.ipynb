```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()

data=[(1, "Alice", 25, "mumbai"),
    (2, "Bob", 30, "karnataka"),
    (3, "Charlie", 35, "noida")]

schema=["id", "name", "age", "city"]

df=spark.createDataFrame(data,schema)

df.show()

df.printSchema()
```
