```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()

df=spark.read.format("csv").option("header","true").option("inferSchema","true").load("/Volumes/data_engineering/raw_data/raw_data/sales_data.csv")

df.select("name","city","sales").display()

df.filter(col("sales")>1500).display()

df.groupby("city").agg(sum("sales").alias("total_sales")).display()
```
