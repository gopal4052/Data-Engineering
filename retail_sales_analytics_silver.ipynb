# Silver ingest

```python
df_orders_bronze = spark.read.format("delta").load("/Volumes/my_local_data/bronze/retail_sales/orders")
df_customers_bronze = spark.read.format("delta").load("/Volumes/my_local_data/bronze/retail_sales/customers")

from pyspark.sql.functions import to_date, col, cast , date_format ,current_date

df_orders_bronze = df_orders_bronze.withColumn("order_date", date_format(col("order_date"), "dd-MM-yyyy"))
df_orders_bronze = df_orders_bronze.withColumn("amount", col("amount").cast("double"))
df_orders_bronze = df_orders_bronze.withColumn("ingestion_date", current_date())

df_customers_bronze = df_customers_bronze.withColumn("signup_date", date_format(col("signup_date"), "dd-MM-yyyy"))
df_customers_bronze = df_customers_bronze.withColumn("ingestion_date", current_date())                                                    

df_orders_bronze.write.format("delta").mode("overwrite").save("/Volumes/my_local_data/silver/retail_sales/orders")
df_customers_bronze.write.format("delta").mode("overwrite").option("overwriteschema",'true').save("/Volumes/my_local_data/silver/retail_sales/customers")

print("Data written to silver successfully")
```
